```python
import sklearn.feature_extraction.text as text #import text module from sklear feature extraction to perform the actions on 
#text data to make the text data into numeric vector form
```

```python
tf=text.TfidfVectorizer(data['product_name'].tolist(),stop_words='english') #here we are using TfidfVectorizer instead of
#countVectorizer since countvecotorizer does not takes the word importance into the account it assigns the same weightage
#to each and every word. and by using the stop words we are ignoring the most commonly used words in english like; to, the 
#hence, a, an etc and also Tfidfvatorizer accepts the text data as in the form of list so we are passing a list.
matrix=tf.fit_transform(data['product_name'].tolist()) #here we are fitting the text data by using fit_transform. 
```

```python
pd.DataFrame(matrix.toarray(),columns=tf.get_feature_names()).head() #if want to see what kind of colums TfidfVectorizer has
#created then use this otherwise ignore.
```

```python
from sklearn.metrics.pairwise import cosine_similarity #in the created vecotor to find out how much two rows are similar we
#use the cosine similarity.
```

```python
sim_unigram=cosine_similarity(matrix) #this will create the cosine similarity for each of the row with respect to other 
#row, then if we want to filter out the number of similar items we can pass the value how many similar items we want to find
#out and can find out easily.
```

```python
def get_similar_items(x):
    return ",".join(data['product_name'].loc[x.argsort()[-5:-1]]) #getting top 5 similar values using the sim_unigram matrix
#which has the similarity matrix for the product_name.
data['Suggested_Products']=[get_similar_items(x) for x in sim_unigram] #use to create the new based based in the given 
#condition in the function.
```

```python
data['product_id']=data['product_id'].map(lambda x: str(x)) #make the product_id columns in string format as we have to
#filter out the top 5 items for each row with the help of cosine similarity
def get_sim_pi(x):
      return ",".join(data['product_id'].loc[x.argsort()[-5:-1]])    
data['suggested_pro_ids']=[get_sim_pi(x) for x in sim_unigram] #use to create the new based based in the given 
#condition in the function.
```

```python
data['department_id']=data['department_id'].map(lambda x: str(x)) #make the product_id columns in string format as we have to
#filter out the top 5 items for each row with the help of cosine similarity
def get_sim_products(x):
    return ",".join(data['department_id'].loc[x.argsort()[-5:-1]])
data['suggested_dept_ids']=[get_sim_products(x) for x in sim_unigram] #use to create the new based based in the given 
#condition in the function.
```
